```python
# load dataset (example: Gutenberg Alice)
file_path = '11-0.txt'   # update path if needed
with open(file_path, 'r', encoding='utf-8') as f:
    text = f.read().lower()
print("Loaded bytes:", len(text))
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
import re
from nltk.tokenize import sent_tokenize, word_tokenize

def clean_text(s):
    s = s.lower()
    s = re.sub(r'[^a-z0-9\s]', ' ', s)   # remove punctuation (basic)
    s = re.sub(r'\s+', ' ', s).strip()
    return s

text_clean = clean_text(text)
sentences = [word_tokenize(sent) for sent in sent_tokenize(text_clean)]
print("Sentences:", len(sentences))
print("Example sentence tokens:", sentences[0][:20])
```

```python
!pip install gensim
```

```python
from gensim.models import Word2Vec

# Train CBOW: sg=0 => CBOW
model = Word2Vec(sentences,
                 vector_size=100,   # embedding size
                 window=5,
                 min_count=2,
                 workers=4,
                 sg=0,
                 epochs=10)

# save model
model.save("cbow_alice.model")
print("Model trained and saved.")
```

```python
# Use the full model object
preds = model.predict_output_word(['the', 'queen', 'of'], topn=5)
print("Predicted center words (CBOW):", preds)
```

```python
preds = model.predict_output_word(['the','queen','of'], topn=8)
for word, prob in preds:
    print(f"{word:15s}  {prob*100:.3f}%")
```
